## Discussion

### 1.Summary

##### The project focuses on investigating whether player characteristics and gaming behavior can predict newsletter subscription rates.  

##### We applied the K-Nearest Neighbors classifier to predict subscription status. This model achieved approximately 71% accuracy on the test set. While this accuracy surpasses the baseline model, its predictive capability remains limited.

##### Moreover, the confusion matrix shows the model can identify most subscribers successfully. However, it has difficulties to accurately classifying non-subscribers due to class imbalance. Furthermore, the ROC curve does not exhibit a high area under the curve, only slightly above the random chance line. This further shows that its ability to distinguish between the two classes is limited.




### 2.Expaction


##### Overall, the results generally meet expectations. Before the analysis, it was reasonable to hypothesize that players with longer playtime or higher experience levels would be more likely to subscribe to the newsletter. The actual results support this hypothesis: subscribers generally have longer playtime, and the subscription rate is slightly higher in the more experienced group. 
##### However, the predictive modeling results were weaker than expected. The behavioral and characteristics variables should have stronger predictive power, but the final model achieved only 71% accuracy. 

### 3.Impact

##### These findings have implications for managing recruitment and engagement in the Minecraft study. Although the model has limited predictive power, it still indicates that players with higher engagement or more experience are slightly more likely to subscribe. This suggests recruitment should focus on players who have shown sustained activity or longer playtime.

### 4.Future Question

##### These findings provide direction for improving future model development. The model in this study did not perform optimally. There are several limitations that may have impacted model accuracy. 
##### First, the dataset exhibits imbalance. The number of players subscribing to newsletters far exceeds those who do not. This imbalance makes it difficult for the model to learn effective patterns related to the minority class, leading to incorrect classification of non-subscribers. Second, there is a limited amount of available training data. Due to the small player base and small behavioral variables, the model lacks sufficient information to distinguish between the two groups. Third, the existing variables are insufficient, and some external factors likely influence players' subscription choices. These missing factors may significantly impact subscription decisions.

##### Addressing these issues requires collecting more comprehensive data and incorporating a broader range of variables. These supplements could help optimize the model and improve prediction accuracy.
